```python
import tensorflow as tf
import numpy as np

### Datasets
in_train = np.array([[-1, -1],
                     [-1,  1],
                     [ 1, -1],
                     [ 1,  1]], dtype=np.float32)

in_test  = in_train.copy()

gt_train = np.array([[0],
                     [1],
                     [1],
                     [0]], dtype=np.float32)
gt_test  = gt_train.copy()

# print(in_train.shape)  ### shape=(4, 2)
# print(gt_train.shape)  ### shape=(4, 1)

### Model
'''
1
    model = tf.keras.models.Sequential([...model Layer])

2
    model = tf.keras.models.Sequential()
    model.add(Layer)
    model.add(Layer)
    ...
 tf.keras.layers.Input(shape=(...))   
'''
### 1
# model = tf.keras.models.Sequential([
#     tf.keras.layers.Input(shape=(2)),
#     tf.keras.layers.Dense(2, activation=tf.keras.activations.sigmoid),
#     tf.keras.layers.Dense(1)
# ])

### 2
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(2)))
model.add(tf.keras.layers.Dense(2, activation=tf.keras.activations.sigmoid))
model.add(tf.keras.layers.Dense(1))

### LossOptimizer
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.5),
              loss=tf.keras.losses.MSE,
              metrics=["accuracy"])
### 
model.fit(in_train, gt_train, batch_size=4, epochs=500)

###  NN  100% model.compile(metric=[]) 
model.evaluate(in_test, gt_test)

###  Return 
raw_predict = model.predict(in_test)  ### , shape=(200, 4)
print("raw_predict:", raw_predict)

###   > 0.5 1 0
lab_predict = np.where(raw_predict > 0.5, 1, 0)  ### np.where()  https://numpy.org/doc/stable/reference/generated/numpy.where.html
print("lab_predict", lab_predict)
```

```python
import numpy as np


def Labels_to_one_hots(labels, category_amount):
    ''' MNIST Dataset Dataset   Dataset'''
    base_one_hot = np.eye(category_amount)  ### np.eye https://numpy.org/devdocs/reference/generated/numpy.eye.html print
    # print("base_one_hot=", base_one_hot)
    one_hots = base_one_hot[labels]         ### print
    # print("labels (just showing first five data):", labels[:5])
    # print("one_hots (just showing first five data):", one_hots[:5])
    # print("")
    one_hots = one_hots.astype(np.float32)  ### 
    return one_hots

####################################################################################################################################################################
'''  '''
def Binary_class_Calculate_accuracy(model, in_test, gt_test):
    '''   gt  '''
    raw_pred = model.predict(in_test)          ###  
    lab_pred = np.where(raw_pred > 0.5, 1, 0)  ###   > 0.5 1 0 np.where()  https://numpy.org/doc/stable/reference/generated/numpy.where.html
    accuracy = _accuracy(lab_pred=lab_pred, lab_gt=gt_test)
    return accuracy

def Multi_class_Calculate_accuracy_by_one_hot_gt(model, in_test, gt_test):
    '''  one-hot gt one-hot'''
    raw_pred = model.predict(in_test)             ###  one-hot vector
    lab_pred = raw_pred.argmax(axis=1)            ### predone hotlabel  index index 
    lab_gt   = gt_test .argmax(axis=1)            ###     gt  one hotlabel  index index 
    accuracy = _accuracy(lab_pred=lab_pred, lab_gt=lab_gt)
    return accuracy

def Multi_class_Calculate_accuracy_by_label_gt(model, in_test, gt_test):
    '''  one-hot gt label '''
    raw_pred = model.predict(in_test)             ###  one-hot vector
    lab_pred = raw_pred.argmax(axis=1)            ### predone hotlabel  index index 
    accuracy = _accuracy(lab_pred=lab_pred, lab_gt=gt_test)
    return accuracy

def _accuracy(lab_pred, lab_gt):
    ''' labelone-hot  label  pred  gt function'''
    lab_correct = (lab_pred == lab_gt)            ### array predgtarray index pred == gt  arrayTrue  Falselab_correctarray
    lab_correct = lab_correct.astype(np.uint8)    ### True  1, False  0
    lab_correct_amount = lab_correct.sum()        ###  pred
    lab_amount = len(lab_correct)                 ###  pred
    accuracy = lab_correct_amount / lab_amount    ### accuracy
    print(f"accuracy written by yourself={accuracy}, same as model.evaluate()")  ###  model.evaluate()   
    return accuracy
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def Remove_zero_2d(datas):
    for go_r, data in enumerate(datas):
        for go_c, value in enumerate(data):
            if(value == 0):
                sign = np.random.uniform(0, 1)
                if(sign > 0.5): datas[go_r, go_c] =  0.00001
                else:           datas[go_r, go_c] = -0.00001
    return datas

def Get_quadrant_labels(datas):
    labels = []
    for data in datas:
        if  (data[0] > 0 and data[1] > 0): labels.append([1, 0, 0, 0])  ### 
        elif(data[0] < 0 and data[1] > 0): labels.append([0, 1, 0, 0])  ### 
        elif(data[0] < 0 and data[1] < 0): labels.append([0, 0, 1, 0])  ### 
        elif(data[0] > 0 and data[1] < 0): labels.append([0, 0, 0, 1])  ### 
    return np.array(labels, dtype=np.float32)  ###  float float32 


def Visualize_current_pred(model, in_test, current_epoch):
    raw_predict = model.predict(in_test)          ### , shape=(200, 4)
    lab_predict = raw_predict.argmax(axis=1) + 1  ###   index index  shape=(200) +1  1,2,3,4 argmax  index 0,1,2,3 +1 
    in_test_pred_1   = in_test[lab_predict == 1]  ### xy   
    in_test_pred_2   = in_test[lab_predict == 2]  ### xy   
    in_test_pred_3   = in_test[lab_predict == 3]  ### xy   
    in_test_pred_4   = in_test[lab_predict == 4]  ### xy   

    ######################################################################################################
    # 
    fig = plt.gcf()            ### 
    fig.set_size_inches(6, 6)  ###   (, )
    fig.tight_layout()         ###    
    ax = plt.gca()   ### 
    ax.cla()         ### cla  clean ax 
    ax.set_title(f"current_epoch:{current_epoch + 1}")  ### 


    #   x,y https://matplotlib.org/stable/tutorials/text/annotations.html
    arrow_pad = 0.02                           ### 
    ### x
    plt.annotate(text="",                      ### 
                 xy=(1 - arrow_pad, 0.5),      ### 
                 xycoords="axes fraction",     ###  (0, 0)(1, 1)
                 xytext=(0 + arrow_pad, 0.5),  ### 
                 textcoords="axes fraction",   ###  (0, 0)(1, 1)
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))  ### 
    ### y
    plt.annotate(text="", xy=(0.5, 1 - arrow_pad), xycoords="axes fraction", xytext=(0.5, 0 + arrow_pad), textcoords="axes fraction", arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))  ### 

    ### (Quadrant)
    ###    
    ### plt.scatter(in_test[:, 0], in_test[:, 1], c=lab_predict, s=10)

    plt.scatter(in_test_pred_1[:, 0],  ###  x
                in_test_pred_1[:, 1],  ###  y
                s=10,                  ###  
                label="pred_Quad_1")   ###    
    plt.scatter(in_test_pred_2[:, 0], in_test_pred_2[:, 1], s=10, label="pred_Quad_2")  ### 
    plt.scatter(in_test_pred_3[:, 0], in_test_pred_3[:, 1], s=10, label="pred_Quad_3")  ### 
    plt.scatter(in_test_pred_4[:, 0], in_test_pred_4[:, 1], s=10, label="pred_Quad_4")  ### 
    plt.legend( loc='lower right')   ###  
    plt.show()

if(__name__ == "__main__"):
    ### 
    ###  Demo epochs  learning_rate 
    ### 
    epochs = 30
    learning_rate = 0.001

    ### Datasets
    in_train = np.random.uniform(-100, 100, (400, 2))
    in_test  = np.random.uniform(-100, 100, (200, 2))

    ### 
    in_train = Remove_zero_2d(in_train)
    in_test  = Remove_zero_2d(in_test)

    gt_train = Get_quadrant_labels(in_train)
    gt_test  = Get_quadrant_labels(in_test)

    # print(in_train.shape)  ### shape=(400, 2)
    # print(gt_train.shape)  ### shape=(200, 2)

    ### Model
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(2)),
        tf.keras.layers.Dense(128, activation='sigmoid'),
        tf.keras.layers.Dense(4)
    ])


    ### LossOptimizer
    model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate),
                loss='mse',
                metrics=['accuracy'])
    ### 
    plt.ion()  ###   plt  plt.show()
    for epoch in range(epochs):
        model.fit(in_train, gt_train, epochs=1)  ###  for model.fit(in_train, gt_train, epochs=100)epochs100100

        #  
        Visualize_current_pred(model, in_test, epoch)
        plt.pause(0.1)  ### plt.ion()  0.1

    #  NN  98% 
    model.evaluate(in_test, gt_test)

    ###  Return 
    raw_predict = model.predict(in_test)  ### , shape=(200, 4)
    print("raw_predict (just showing first five data):", raw_predict[:5])

    ###   index index  shape=(200) +1  1,2,3,4 argmax  index 0,1,2,3 +1 
    lab_predict = raw_predict.argmax(axis=1) + 1
    print("lab_predict (just showing first five data):", lab_predict[:5])

    ### print  
    ###     predict  OK 
    plt.ioff()  ###  plt
    Visualize_current_pred(model, in_test, epochs)  ###   

    Multi_class_Calculate_accuracy_by_one_hot_gt(model, in_test, gt_test)
```
