```python
import tensorflow as tf
import numpy as np

### Datasets部分
in_train = np.array([[-1, -1],
                     [-1,  1],
                     [ 1, -1],
                     [ 1,  1]], dtype=np.float32)

in_test  = in_train.copy()

gt_train = np.array([[0],
                     [1],
                     [1],
                     [0]], dtype=np.float32)
gt_test  = gt_train.copy()

# print(in_train.shape)  ### shape=(4, 2)
# print(gt_train.shape)  ### shape=(4, 1)

### Model部分
'''
宣告方式1：
    model = tf.keras.models.Sequential([...疊model的 Layer物件放這邊])

宣告方式2：
    model = tf.keras.models.Sequential()
    model.add(Layer物件放這邊)
    model.add(Layer物件放這邊)
    ...
通常第一層都會放 tf.keras.layers.Input(shape=(...)) 來告訴網路 你的輸入資料 長什麼樣子
'''
### 宣告方式1例子：
# model = tf.keras.models.Sequential([
#     tf.keras.layers.Input(shape=(2)),
#     tf.keras.layers.Dense(2, activation=tf.keras.activations.sigmoid),
#     tf.keras.layers.Dense(1)
# ])

### 宣告方式2：
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(2)))
model.add(tf.keras.layers.Dense(2, activation=tf.keras.activations.sigmoid))
model.add(tf.keras.layers.Dense(1))

### Loss、Optimizer部分
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.5),
              loss=tf.keras.losses.MSE,
              metrics=["accuracy"])
### 訓練模型
model.fit(in_train, gt_train, batch_size=4, epochs=500)

### 訓練後的 NN 在測試集上可得到近 100% 正確辨識率，這裡顯示的指標是根據model.compile(metric=[在這裡指定的])， 如果想看別種指標也可以自己設定
model.evaluate(in_test, gt_test)

### 輸入測試資料進網路， Return 網路輸出的結果
raw_predict = model.predict(in_test)  ### 未處理的網路輸出, shape=(200, 4)
print("raw_predict:", raw_predict)

### 對網路輸出作後處理，找出每一筆資料 如果 > 0.5 就指定為1， 否則指定為0
lab_predict = np.where(raw_predict > 0.5, 1, 0)  ### np.where() 用法可參考： https://numpy.org/doc/stable/reference/generated/numpy.where.html
print("lab_predict", lab_predict)
```

```python
import numpy as np


def Labels_to_one_hots(labels, category_amount):
    ''' 這個轉法式剛好可以符合MNIST Dataset ，未來如果有自己設計Dataset 或 用別人的 Dataset轉法也自己注意怎麼轉喔'''
    base_one_hot = np.eye(category_amount)  ### np.eye 可參考：https://numpy.org/devdocs/reference/generated/numpy.eye.html ，如果不懂就自己把值print出來慢慢體會吧～
    # print("base_one_hot=", base_one_hot)
    one_hots = base_one_hot[labels]         ### 如果不懂就自己把值print出來慢慢體會吧～
    # print("labels (just showing first five data):", labels[:5])
    # print("one_hots (just showing first five data):", one_hots[:5])
    # print("")
    one_hots = one_hots.astype(np.float32)  ### 記得轉一下型態
    return one_hots

####################################################################################################################################################################
''' 寫法百百種，為了打註解才寫那麼長，要自己改更好的寫法都可以喔 '''
def Binary_class_Calculate_accuracy(model, in_test, gt_test):
    ''' 網路輸出為 一個值的形式， gt也為 一個值的形式 '''
    raw_pred = model.predict(in_test)          ### 未處理的網路輸出，為 一個值
    lab_pred = np.where(raw_pred > 0.5, 1, 0)  ### 找出每一筆資料 如果 > 0.5 就指定為1， 否則指定為0， np.where() 用法可參考： https://numpy.org/doc/stable/reference/generated/numpy.where.html
    accuracy = _accuracy(lab_pred=lab_pred, lab_gt=gt_test)
    return accuracy

def Multi_class_Calculate_accuracy_by_one_hot_gt(model, in_test, gt_test):
    ''' 網路輸出為 one-hot形式， gt也為 one-hot形式'''
    raw_pred = model.predict(in_test)             ### 未處理的網路輸出，為 one-hot vector
    lab_pred = raw_pred.argmax(axis=1)            ### 網路pred為one hot，要轉成label：找出每一筆資料最大值 的 位置index， 該index值代表 其所屬的類別
    lab_gt   = gt_test .argmax(axis=1)            ###     gt  為one hot，要轉成label：找出每一筆資料最大值 的 位置index， 該index值代表 其所屬的類別
    accuracy = _accuracy(lab_pred=lab_pred, lab_gt=lab_gt)
    return accuracy

def Multi_class_Calculate_accuracy_by_label_gt(model, in_test, gt_test):
    ''' 網路輸出為 one-hot形式， gt為 label形式 '''
    raw_pred = model.predict(in_test)             ### 未處理的網路輸出，為 one-hot vector
    lab_pred = raw_pred.argmax(axis=1)            ### 網路pred為one hot，要轉成label：找出每一筆資料最大值 的 位置index， 該index值代表 其所屬的類別
    accuracy = _accuracy(lab_pred=lab_pred, lab_gt=gt_test)
    return accuracy

def _accuracy(lab_pred, lab_gt):
    ''' 輸入進來的格式都要是label形式，不能是one-hot喔， 只要都轉成 label 且 pred 和 gt 結構一樣就可以用此function'''
    lab_correct = (lab_pred == lab_gt)            ### 建立一個新array，內部的值是看 pred和gt兩個array中 一樣index的位置如果 pred值 == gt值 則 新array該位置設為True， 否則 False，最後用lab_correct名字指向新array
    lab_correct = lab_correct.astype(np.uint8)    ### True 變 1, False 變 0
    lab_correct_amount = lab_correct.sum()        ### 總共有幾筆 pred是對的
    lab_amount = len(lab_correct)                 ### 總共有幾筆 pred
    accuracy = lab_correct_amount / lab_amount    ### 計算accuracy
    print(f"accuracy written by yourself={accuracy}, same as model.evaluate()")  ### 雖然值是跟 model.evaluate()一樣， 不過我不知道我的作法是不是跟 他內部寫的一樣喔， 這邊只是給大家參考有個感覺而已～
    return accuracy
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def Remove_zero_2d(datas):
    for go_r, data in enumerate(datas):
        for go_c, value in enumerate(data):
            if(value == 0):
                sign = np.random.uniform(0, 1)
                if(sign > 0.5): datas[go_r, go_c] =  0.00001
                else:           datas[go_r, go_c] = -0.00001
    return datas

def Get_quadrant_labels(datas):
    labels = []
    for data in datas:
        if  (data[0] > 0 and data[1] > 0): labels.append([1, 0, 0, 0])  ### 第一象限
        elif(data[0] < 0 and data[1] > 0): labels.append([0, 1, 0, 0])  ### 第二象限
        elif(data[0] < 0 and data[1] < 0): labels.append([0, 0, 1, 0])  ### 第三象限
        elif(data[0] > 0 and data[1] < 0): labels.append([0, 0, 0, 1])  ### 第四象限
    return np.array(labels, dtype=np.float32)  ### 因為網路訓練 的過程都是float運算，所以建議轉成 float32， 否則容易出錯


def Visualize_current_pred(model, in_test, current_epoch):
    raw_predict = model.predict(in_test)          ### 未處理的網路輸出, shape=(200, 4)
    lab_predict = raw_predict.argmax(axis=1) + 1  ### 對網路輸出作後處理，找出每一筆資料最大值 的 位置index， 該index值代表 該筆資料其所屬的類別， 結果的shape=(200)， +1 是因為象限是 1,2,3,4， argmax 回傳的是 index 0,1,2,3， +1 才可以跟象限對上
    in_test_pred_1   = in_test[lab_predict == 1]  ### 輸入的xy裡， 目前網路判斷為 第一象限 的點有哪些
    in_test_pred_2   = in_test[lab_predict == 2]  ### 輸入的xy裡， 目前網路判斷為 第二象限 的點有哪些
    in_test_pred_3   = in_test[lab_predict == 3]  ### 輸入的xy裡， 目前網路判斷為 第三象限 的點有哪些
    in_test_pred_4   = in_test[lab_predict == 4]  ### 輸入的xy裡， 目前網路判斷為 第四象限 的點有哪些

    ######################################################################################################
    # 開始畫圖
    fig = plt.gcf()            ### 拿到目前正在使用的畫框
    fig.set_size_inches(6, 6)  ### 調整畫框大小， 設定方法是 (寬, 高)
    fig.tight_layout()         ### 畫框邊緣 跟 畫布邊緣 要緊密的靠近
    ax = plt.gca()   ### 拿到目前畫框內正在使用的畫布
    ax.cla()         ### 把畫布的內容清空，cla 是 clean ax 的縮寫吧我猜
    ax.set_title(f"current_epoch:{current_epoch + 1}")  ### 設定畫布標題


    # 用畫建頭的方式 來畫 x,y軸， 如何畫箭頭可參考：https://matplotlib.org/stable/tutorials/text/annotations.html
    arrow_pad = 0.02                           ### 箭頭到邊緣之間的距離
    ### 畫x軸
    plt.annotate(text="",                      ### 建尾要顯示的字
                 xy=(1 - arrow_pad, 0.5),      ### 箭頭在哪
                 xycoords="axes fraction",     ### 坐標系用 畫布左下角當(0, 0)，右上角當(1, 1)
                 xytext=(0 + arrow_pad, 0.5),  ### 箭尾在哪
                 textcoords="axes fraction",   ### 坐標系用 畫布左下角當(0, 0)，右上角當(1, 1)
                 arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))  ### 畫箭頭的樣式
    ### 畫y軸
    plt.annotate(text="", xy=(0.5, 1 - arrow_pad), xycoords="axes fraction", xytext=(0.5, 0 + arrow_pad), textcoords="axes fraction", arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))  ### 同上道裡

    ### 畫各個象限(Quadrant)的點
    ### 各組分別出來畫才可以 用 各顏色標示圖 喔～　
    ### 要不然一行就夠了：plt.scatter(in_test[:, 0], in_test[:, 1], c=lab_predict, s=10)

    plt.scatter(in_test_pred_1[:, 0],  ### 點的 x
                in_test_pred_1[:, 1],  ### 點的 y
                s=10,                  ### 點的 大小
                label="pred_Quad_1")   ### 點的 標籤， 有指定標籤才有圖中右下角的 各顏色標示圖喔～
    plt.scatter(in_test_pred_2[:, 0], in_test_pred_2[:, 1], s=10, label="pred_Quad_2")  ### 同上理
    plt.scatter(in_test_pred_3[:, 0], in_test_pred_3[:, 1], s=10, label="pred_Quad_3")  ### 同上理
    plt.scatter(in_test_pred_4[:, 0], in_test_pred_4[:, 1], s=10, label="pred_Quad_4")  ### 同上理
    plt.legend( loc='lower right')   ### 各顏色標示圖 要顯示在哪邊
    plt.show()

if(__name__ == "__main__"):
    ### 設定參數部分
    ### 為了 Demo視覺化時不要調整太快 epochs才調那麼少 和 learning_rate 才調那麼小
    ### 可自行嘗試怎麼調更好喔～
    epochs = 30
    learning_rate = 0.001

    ### Datasets部分
    in_train = np.random.uniform(-100, 100, (400, 2))
    in_test  = np.random.uniform(-100, 100, (200, 2))

    ### 資料做前處理
    in_train = Remove_zero_2d(in_train)
    in_test  = Remove_zero_2d(in_test)

    gt_train = Get_quadrant_labels(in_train)
    gt_test  = Get_quadrant_labels(in_test)

    # print(in_train.shape)  ### shape=(400, 2)
    # print(gt_train.shape)  ### shape=(200, 2)

    ### Model部分
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(2)),
        tf.keras.layers.Dense(128, activation='sigmoid'),
        tf.keras.layers.Dense(4)
    ])


    ### Loss、Optimizer部分
    model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate),
                loss='mse',
                metrics=['accuracy'])
    ### 訓練模型
    plt.ion()  ### 視覺化的東西， 開啟 plt互動模式， 意思是 plt.show()時程是不會停下來會繼續跑
    for epoch in range(epochs):
        model.fit(in_train, gt_train, epochs=1)  ### 為了視覺化才寫成這樣， 要不然其實整個for迴圈可濃縮成 model.fit(in_train, gt_train, epochs=100)，epochs指定100即可訓練100次

        # 視覺化 訓練過程中的結果
        Visualize_current_pred(model, in_test, epoch)
        plt.pause(0.1)  ### 因為plt.ion() 程式不會停下來， 所以要自行暫停0.1秒來看圖

    # 訓練後的 NN 在測試集上可得到近 98% 正確辨識率
    model.evaluate(in_test, gt_test)

    ### 輸入測試資料進網路， Return 網路輸出的結果
    raw_predict = model.predict(in_test)  ### 未處理的網路輸出, shape=(200, 4)
    print("raw_predict (just showing first five data):", raw_predict[:5])

    ### 對網路輸出作後處理，找出每一筆資料最大值 的 位置index， 該index值代表 該筆資料其所屬的類別， 結果的shape=(200)， +1 是因為象限是 1,2,3,4， argmax 回傳的是 index 0,1,2,3， +1 才可以跟象限對上
    lab_predict = raw_predict.argmax(axis=1) + 1
    print("lab_predict (just showing first five data):", lab_predict[:5])

    ### 以上只是print出 模型輸出各點的類別， 但因為沒有搭配輸入一起看也不知道對不對
    ### 因此需要更進一步 視覺化 訓練完的結果， 才能更好的理解 predict 出來的東西 是不是OK 的
    plt.ioff()  ### 關閉 plt的互動模式
    Visualize_current_pred(model, in_test, epochs)  ### 視覺化  訓練過程後的結果

    Multi_class_Calculate_accuracy_by_one_hot_gt(model, in_test, gt_test)
```
